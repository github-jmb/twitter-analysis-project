```python
import pandas as pd
```

```python
df = pd.read_csv('Tweets_Overall.csv')
df.shape
```

```python
df.sample(n=10)
```

```python
df['club'].value_counts()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
dfLiverpool = df.loc[df['club']=='Liverpool FC']

fig = make_subplots(rows=1, cols=1,)

fig.add_trace(go.Scatter(x=dfLiverpool['createdAt'], 
        y=dfLiverpool['favorites'],
        name="Liverpool FC Favorites"),row=1, col=1)

fig.update_layout(height=500, width=700,
        title_text="Premier League Twitter Analysis - Favorite Count")

fig.show()
```

```python
dfLiverpool = df.loc[df['club']=='Liverpool FC']
dfManU = df.loc[df['club']=='Manchester United']
dfArsenal = df.loc[df['club']=='Arsenal']
dfChelsea = df.loc[df['club']=='Chelsea']
dfManCity = df.loc[df['club']=='Manchester City']
dfSpurs = df.loc[df['club']=='Tottenham']
```

```python
fig = make_subplots(
    rows=6, cols=1
    )

fig.add_trace(go.Scatter(x=dfLiverpool['createdAt'], y=dfLiverpool['favorites'],
              name="Liverpool FC Favorites"),
              row=1, col=1)

fig.add_trace(go.Scatter(x=dfManU['createdAt'], y=dfManU['favorites'],
              name="Manchester United Favorites"),
              row=2, col=1)

fig.add_trace(go.Scatter(x=dfArsenal['createdAt'], y=dfArsenal['favorites'],
              name="Arsenal Favorites"),
              row=3, col=1)

fig.add_trace(go.Scatter(x=dfManCity['createdAt'], y=dfManCity['favorites'],
              name="Manchester City Favorites"),
              row=4, col=1)

fig.add_trace(go.Scatter(x=dfSpurs['createdAt'], y=dfSpurs['favorites'],
              name="Tottenham Hotspur Favorites"),
              row=5, col=1)

fig.add_trace(go.Scatter(x=dfChelsea['createdAt'], y=dfChelsea['favorites'],
              name="Chelsea Favorites"),
              row=6, col=1)

fig.update_layout(height=500, width=700,
                  title_text="Premier League Twitter Analysis - Favorite Count")

fig.show()
```
